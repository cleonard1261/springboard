```python
import json
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

with open("logins.json") as login_file:
    json_login = json.load(login_file)
print(len(json_login['login_time']))
df = pd.DataFrame(json_login['login_time'],columns=['login_time_str'] )

# change from string to datetime. This should test that dates are valid.
df['login_time'] = [datetime.strptime(dt, '%Y-%m-%d %H:%M:%S') for dt in json_login['login_time']]
print(df.login_time.describe())

# drop duplicate records
df = df.sort_values(by='login_time').reset_index().drop(columns=['index']).drop_duplicates()
print(len(df))
print(df.login_time.describe())
df.head()
```

```python
def get_time_interval(d):
    """ 
    Passes in the login time as a string in its datetime format. Parses out the hour and minute from each date
    and determines which 15 minute interval within each hour that the login occurred. Returns a string that represents
    the hour and 15 minute interval. The 15 minute intervals are (inclusive):
     0 to 14 ==> 14
    15 to 29 ==> 29
    30 to 44 ==> 44
    45 to 59 ==> 59
    """
    d = d.split(' ')[1]
    h = d[:2]
    m = int(d[3:5])
    if m < 15:
        return h + ':14'
    elif m < 30:
        return h + ':29'
    elif m < 45:
        return h + ':44'
    elif m < 60:
        return h + ':59'
    else:
        print(str(m) + 'is not valid')

df['time_interval'] = df.login_time_str.apply(get_time_interval)
df['day_of_week'] = df.login_time.dt.weekday
df.head()
```

```python
df = df.sort_values(by='time_interval')
df.head()
```

```python
print(str("                  Login counts per 15 minute time interval per day of the week.").upper())
print('='*117)
print('     Monday         Tuesday          Wednesday         Thursday          Friday          Saturday         Sunday')
_ = sns.catplot(x="time_interval", col="day_of_week",data=df, kind="count",
                 height=20, aspect=.8)
```

   *Ultimate Technologies Inc. is an American worldwide online transportation network company that has disrupted the taxi and logistics industry and is a prestigious companies to work at. This challenge has been adapted from an actual Ultimate Inc. data science challenge.*
   
   
   <br/>
   
   
   From the description of the Ultimate Technologies Inc. provided by the company itself, presumably, it is unclear to me what the "logistics" industry refers to. It sounds like it could be delivery services of some sort, possibly similar to UPS or Uber Eats. 
  
  
   <br/>
  
   We can assume that the taxi services are similar to the services provided by Uber or Lyft. It is those services that seem to have the greatest effect on the patterns seen in the graphs above.
   
   <br/>
  
  There are 96 fifteen minute time intervals for each day. That makes sense because there are 24 hours a day and 4 fifteen minute segments in each hour. Surprisingly, there are logins recorded for every 15 minute interval for every day of the week aggregated by the day of the week. No 15 minute interval recorded 0 logins. 
  

 
 
  <br/>
 
  The graphs above represent the login counts for every 15 minute period for the entire day for every day of the week. It's apparent that as the week goes on the counts start to increase. All the business days seems to follow the same pattern, albeit increasing as the week goes on. It looks like there is some light traffic in the early hours of the day then as the normal work day starts it settles down to just about nothing. Then there's some movement around lunch. Most likely some people Ultimating to lunch or ordering food for delivery. Then the afternoon is tame until the end of the work day and things pick up some again. Work day patterns are very similar.
  
  
   <br/>
  
  
  Not surprisingly, the weekend patterns are different. Starting on Friday night volume increases and stays pretty steady until about breakfast time on Saturday. Then it's quite slow until the afternoon and then really picks up in the evening. The odd thing about Saturday evening/night is that the moderate increase is much less than the increase for early Sunday morning. It would be natural to assume the same sort of pattern as that of Fright night/Saturday morning.
